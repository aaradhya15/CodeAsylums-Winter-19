```python

def calculateEuclideanDistance(trainRow,testRow):
    sum=0
    for i in range(len(trainRow)-1):
        sum = sum+(trainRow[i]-testRow[i])*(trainRow[i]-testRow[i])
        #print(sum)
    return math.sqrt(sum)
        
    
```

```python
df[['sepal_length','sepal_width','petal_length','petal_width']] = df[['sepal_length','sepal_width','petal_length','petal_width']].apply(pd.to_numeric)
dfCopy=df.copy()
trainSet = dfCopy.sample(frac=0.60, random_state=0)
validateSet = dfCopy.sample(frac=0.20, random_state=0)
testSet = dfCopy.sample(frac=0.20, random_state=0)
#print(trainSet,validateSet,testSet)
#print(len(trainSet),len(validateSet),len(testSet))


def validate(trainSet,validateRow,k):
    distances=list()
    for i in range(len(trainSet)):
        distance = calculateEuclideanDistance(trainSet.iloc[i],validateRow)
        distances.append((distance,trainSet.iloc[i]))
    distances.sort(key = lambda x: x[0])
    neighbours=list()
    for i in range(k):
        neighbours.append(distances[i][1])
    outputs = list()
    for row in neighbours:
        outputs.append(row[-1])
    prediction = max(set(outputs))
    return prediction
        
        

print(validateSet.iloc[0])
validate(trainSet,validateSet.iloc[0],3)
        
        
    
```
